```
def Rebalance(c,emp1,emp2):
  c.index.to_pydatetime()
  for x in range(2012,2016):
    f1=c[c.index.year==x]
    f1.index.to_pydatetime()
    for y in range(1,13):
      f2=f1[f1.index.month==y]
      emp1.append(f2.values.tolist()[-1])

  for x in range(2016,2020):
    f1=c[c.index.year==x]
    f1.index.to_pydatetime()
    for y in range(1,13):
      f2=f1[f1.index.month==y]
      emp2.append(f2.values.tolist()[-1])
```

Loss Function
$$
L= -[V_{t+1}-V_t] =-\sum\lambda_i(S_{t+1,i}-S_{t,i})=-\sum\lambda_iS_{t_i}(e^{X_{t+1,i}}-1)
$$
Linearized Loss Function
$$
e^x=1+x \\
L^{\Delta}_{t+1}=-\sum\lambda_iS_{t,i}X_{t+1,i}
$$
We have $X_{t+1,i}$ is normal distributed in $N(\mu, \sigma^2)$
